```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import csv,os
from pathlib import Path
import plotly.express as px
import pandas as pd
import plotly.graph_objs as go
import math
#from app import app

from dash.dependencies import Input, Output
```

```python
def return_abs_path(file_name):
    data_folder = "apartment-weather/" + file_name
    relative_file_path = Path(data_folder)
    absolute_file_path = relative_file_path.absolute()
    return absolute_file_path

def return_abs_path2(file_name):
    data_folder = "apartment/" + file_name
    relative_file_path = Path(data_folder)
    absolute_file_path = relative_file_path.absolute()
    return absolute_file_path
```

```python
df1_2016 = pd.read_csv(return_abs_path("apartment2016.csv"))
d = pd.to_datetime(df1_2016["time"],unit='s')
df1_2016["date_and_time"] = d.dt.tz_localize("GMT").dt.tz_convert('America/New_York').dt.tz_localize(None)
df1_2016["date"] = pd.to_datetime(df1_2016["date_and_time"]).apply(lambda x: x.date())
#print(df1["date"]) # Converting to date and time


df1_2015 = pd.read_csv(return_abs_path("apartment2015.csv"))
d = pd.to_datetime(df1_2015["time"],unit='s')
df1_2015["date_and_time"] = d.dt.tz_localize("GMT").dt.tz_convert('America/New_York').dt.tz_localize(None)
df1_2015["date"] = pd.to_datetime(df1_2015["date_and_time"]).apply(lambda x: x.date())
#print(df1["date"]) # Converting to date and time


df1_2014 = pd.read_csv(return_abs_path("apartment2014.csv"))
d = pd.to_datetime(df1_2014["time"],unit='s')
df1_2014["date_and_time"] = d.dt.tz_localize("GMT").dt.tz_convert('America/New_York').dt.tz_localize(None)
df1_2014["date"] = pd.to_datetime(df1_2014["date_and_time"]).apply(lambda x: x.date())
#print(df1["date"]) # Converting to date and time
df1 = df1_2016.append(df1_2015,ignore_index=True)
df1 = df1.append(df1_2014,ignore_index=True) # contains all the weather data
#print(df1.shape)
```

```python
df2 = df1.groupby("date")
df2_x = df2.apply(lambda grp: round(grp.mean(),2))
df2_x.to_csv("Apt_temp.csv", sep=',',index=True, header=False)
df2_x = pd.read_csv("Apt_temp.csv",names=['date','temperature', 'humidity', 'visibility', 'apparentTemperature', 'pressure', 'windSpeed', 'cloudCover', 'time', 'windBearing', 'precipIntensity', 'dewPoint', 'precipProbability'])
#contains weather data grouped by date
```

```python
df_main = pd.DataFrame(columns=['power'])
for year in range(2014,2017):
    for i in range(1,115):
        name = str(year)+"/Apt"+str(i)+"_"+str(year)+".csv"
        try:
            df3 = pd.read_csv(return_abs_path2(name),names=['date_and_time', 'power'], header=None)
            df3["date"] = pd.to_datetime(df3["date_and_time"]).apply(lambda x: x.date())
            df4 = df3.groupby("date")
            df4_x = df4.apply(lambda grp: round(grp.mean(),2))
            df_main = df_main.append(df4_x)
        except FileNotFoundError:
            continue
    

df_main.to_csv("Apt_temp.csv", sep=',',index=True, header=False)
df_main = pd.read_csv("Apt_temp.csv",names=["date", "power"])
df_main = df_main.groupby("date")
df_main = df_main.apply(lambda grp: round(grp.mean(),2))
df_main.to_csv("Apt_temp.csv", sep=',',index=True, header=False)
df_main = pd.read_csv("Apt_temp.csv",names=["date", "power"])
```

```python
df3 = df_main.merge(df2_x,on='date',how='inner')
#print(df3)
```

```python
df4 = df3.loc[:,['temperature','power']] #df3 is the full dataframe
df4 = df4.groupby('temperature')
df4 = df4.apply(lambda grp: round(grp.mean(),2))
df4.to_csv("Apt_temp.csv", sep=',',index=False, header=False)
df4 = pd.read_csv("Apt_temp.csv",names=['temperature','power'])
#print(df4)
```

```python
fig = px.line(df4, x='temperature', y='power', title='Mean Power Consumption across different temperatures(2014-2016)')
fig.update_xaxes(ticks="inside",title_font=dict(family='Georgia', color='black'),tickfont=dict(family='Georgia', color='black'))
fig.update_yaxes(title_font=dict(family='Georgia', color='black'),tickfont=dict(family='Georgia', color='black'))
fig.update_layout(font=dict(family="Georgia",color="black"),xaxis_title="Temperature(in F)", yaxis_title="Power(kW)")
fig.show()
```

```python
df8 = df1.loc[:,['summary','date']] #extracting summary and date from df1; the weather unsummarized data
df8 = df8.groupby("date")['summary'].agg(lambda x: pd.Series.mode(x)[0]).to_frame() # selecting the most frequent weather summary
df8.to_csv("Apt_temp.csv", sep=',',index=True, header=False)
df8 = pd.read_csv("Apt_temp.csv",names=["date", "summary"])
df9 = df_main.merge(df8,on='date',how='inner') #contains date, power, summary
```

```python
df10 = df9.loc[:,['summary','power']] 
df10 = df10.groupby('summary')
df10 = df10.apply(lambda grp: round(grp.mean(),2))
df10.to_csv("Apt_temp.csv", sep=',',index=True, header=False)
df10 = pd.read_csv("Apt_temp.csv",names=['summary','power'])
```

```python
fig = px.bar(df10, x='summary', y='power', title='Mean Power Consumption across different weather summaries(2014-2016)', text='power')
fig.update_xaxes(ticks="inside",title_font=dict(family='Georgia', color='black'),tickfont=dict(family='Georgia', color='black'))
fig.update_yaxes(title_font=dict(family='Georgia', color='black'),tickfont=dict(family='Georgia', color='black'))
fig.update_layout(font=dict(family="Georgia",color="black"),xaxis_title="Weather Summary", yaxis_title="Power(kW)")
fig.show()
```
